# Github Pages

### Purpose: 
allow you to post your pages, and let Github handle the rest


### Turn on and Deploy: 
github repository settings -> 左边目录：Pages

Branch -> pick the main branch -> Select the /root folder -> save

Once done, github will give you an URL
- 种类：html page
- 默认post: .readme file
- 页面内容：main branch中的东西都可access；可输入层级进入 i.e. url/file1/file1_1/...


### Created by: github action
go to: Action （repo横目录） -> Action -> check the page-created workflow

## create new github action

### purpose: 
using github actions to run the py.file and push it to the server(current repo)

### how:
go to action -> workflow -> create a new workflow -> set up a workflow yourself

### params:


name: action name
on: on command --start processing
- push: 本action执行发生于：每当有人push a change onto the main branch


job: what you want to do
1. build
    - runs-on: run it on a specific type of machine 定义要跑的机器型号

    - steps: 具体how it works的步骤（--也可以seasrch右侧列出的 -> marketplace actions）
    - i.e. 以下为截图中的例子
        - name: 步骤名称（以这一步要干什么命名）
        - uses: actions/要执行的action种类
            - @v3: 只check version3 --防止别人提交错误文档 i.e. bad v4
            - actions/setup-python@4 --to make sure github to set up python in order to run the .py file
            - with: add other varibal and param for the action -->python版本可以加在这里
        - run: == run command you'll type in the Ternminal --要执行的命令行
            - upgrade pip
            - install pyyaml package
            - or run the file: feed.py
        - name: Push Repo
            - your user name
            - your email
            - git add. -- add all files that generated after running the .py
            - git commit: leave a message
            - git push


### hit 'commit changes'
make sure you give all the permission: 
- repo settings -> left table - Actions -> general -> Workflow permission (the last one): Read and write permissions

如果action errors(执行失败)，更改后重新执行需要在文件中做一处修改（任一修改即可）然后重新执行pushing to main.branch
- 最简单：edit sth in .readme
